## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import numpy as np

# Optimización de la simulación usando NumPy
def lanzar_dados(n_simulaciones=1000):
    # Simular lanzamientos de dados
    dados = np.random.randint(1, 7, (n_simulaciones, 2))  # Simula 1000 pares de dados a la vez
    
    # Sumar los dos dados para cada simulación
    suma_dados = np.sum(dados, axis=1)  # Sumar a lo largo de las columnas (cada fila es una simulación)
    
    # Condición: suma mayor a 7 o número par
    condicion = (suma_dados > 7) | (suma_dados % 2 == 0)
    
    # Contar cuántas veces se cumple la condición
    exitos = np.sum(condicion)
    
    # Calcular probabilidad
    probabilidad = exitos / n_simulaciones
    
    # Mostrar el resultado de forma ordenada
    return f"La probabilidad de obtener una suma mayor a 7 o un número par es: {np.round(probabilidad * 100, 2)}%"

print(lanzar_dados())
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
import random
from collections import Counter

# Inicialización de la caja con las bolas
ball_box = ["White"] * 10 + ["Red"] * 20 + ["Green"] * 30

# Variables para contar los resultados
count_3_white_2_red = 0
count_same_color = 0

# Ejecutar 1000 veces el experimento
for _ in range(1000):
    # Selección de 5 bolas con reemplazo
    selected_balls = [random.choice(ball_box) for _ in range(5)]
    
    # Contar cuántas bolas de cada color se seleccionaron
    ball_counts = Counter(selected_balls)
    
    # Comprobar la condición de 3 blancas y 2 rojas
    if ball_counts["White"] == 3 and ball_counts["Red"] == 2:
        count_3_white_2_red += 1
    
    # Comprobar si todas las bolas son del mismo color
    if len(ball_counts) == 1:  # Solo hay un tipo de bola en la selección
        count_same_color += 1

# Calcular las probabilidades
prob_3_white_2_red = count_3_white_2_red / 1000
prob_same_color = count_same_color / 1000

# Resultados en español
print(f"Probabilidad de tomar 3 bolas blancas y 2 rojas: {prob_3_white_2_red:.4f}")
print(f"Probabilidad de tomar 5 bolas del mismo color: {prob_same_color:.4f}")
```
